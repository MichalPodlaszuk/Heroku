```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data.csv', encoding = "ISO-8859-1")
```

```python
df.Awards.str.split(",").tolist() 
```

```python
df["Awards"] = df["Awards"].str.split(",").tolist()
df.Awards.str.len()
```

```python
def tran_awards(x):
    if x == x:
        return len(x)
    else:
        pass
```

```python
df["Awards"] = df["Awards"].apply(tran_awards) 
```
